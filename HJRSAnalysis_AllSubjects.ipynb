```python
import seaborn as sns
import matplotlib.pyplot as plt
# df_X = df_CSSubjects[['Scopus_Quartile', 'SJR Quartile', 'HJRS Category', 'ESA_Quartile', 'ESA_Quartile_Imputed']]
df_X = df_CSSubjects[['Scopus_Quartile', 'SJR_Quartile', 'HJRS_Category']]
df_Y = df_CSSubjects[['Scopus_Quartile', 'SJR_Quartile', 'ESA_Quartile']]
df_Z = df_CSSubjects[['Scopus_Quartile', 'SJR_Quartile', 'ESA_Quartile_Imputed']]

sjr_quartiles = df_X['SJR_Quartile']
scopus_quartiles = df_X['Scopus_Quartile']
hjrs_labels = df_X['HJRS_Category']

# Generate pivot table
pivot_table1 = pd.pivot_table(df_X, index=[scopus_quartiles, sjr_quartiles], columns=hjrs_labels, aggfunc=len, fill_value=0)

# Generate a light color palette
light_palette = sns.color_palette("light:#5A9", as_cmap=True)

# Plot heatmap with the light color palette
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table1, cmap=light_palette, annot=True, fmt='d', linewidths=0.5, linecolor='black', cbar=False)
plt.title('Pivot Table Heatmap of SJR Quartiles and Scopus Quartiles with HJRS Labels')
plt.xlabel('HJRS Labels')
plt.ylabel('SJR and Scopus Quartiles')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sjr_quartiles = df_Y['SJR_Quartile']
scopus_quartiles = df_Y['Scopus_Quartile']
ESA_labels = df_Y['ESA_Quartile']

# Generate pivot table
pivot_table2 = pd.pivot_table(df_Y, index=[scopus_quartiles, sjr_quartiles], columns=ESA_labels, aggfunc=len, fill_value=0)

# Display the pivot table
print("Pivot Table of SJR Quartiles and Scopus Quartiles with ESA Labels:")
print(pivot_table2)

# Generate a light color palette with shades of green
light_palette = sns.color_palette("light:green", as_cmap=True)

# Plot heatmap with the light color palette
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table2, cmap=light_palette, annot=True, fmt='d', linewidths=0.5, linecolor='black', cbar=False)
plt.title('Pivot Table Heatmap of SJR Quartiles and Scopus Quartiles with ESA Index')
plt.xlabel('ESA Index')
plt.ylabel('SJR and Scopus Quartiles')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sjr_quartiles = df_Z['SJR_Quartile']
scopus_quartiles = df_Z['Scopus_Quartile']
ESA_Quartile_Imputed_labels = df_Z['ESA_Quartile_Imputed']

# Generate pivot table
pivot_table3 = pd.pivot_table(df_Z, index=[scopus_quartiles, sjr_quartiles], columns=ESA_Quartile_Imputed_labels, aggfunc=len, fill_value=0)

# Display the pivot table
print("Pivot Table of SJR Quartiles and Scopus Quartiles with ESA Index (imputed):")
print(pivot_table3)

sns.set_style("white")

# Plot heatmap with the 'Paired' colormap and no color bar
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table3, cmap='Paired', annot=True, fmt='d', linewidths=0.5, linecolor='black', cbar=False)
plt.title('Pivot Table Heatmap of SJR Quartiles and Scopus Quartiles with ESA Index (imputed)')
plt.xlabel('ESA Index(imputed)')
plt.ylabel('SJR and Scopus Quartiles')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
